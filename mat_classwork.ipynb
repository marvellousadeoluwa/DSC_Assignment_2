```python
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x **2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Exercise 1
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
```

```python
#Exercise 2
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,"r")
ax2.plot(x,y,"r")



#Setting Limit
#ax.set_limit(20,50)
#ax.set_llimit(0,50)
```
